Olist order items dataset cleaning

```python
import pandas as pd

df_order_items = pd.read_csv('olist_order_items_dataset.csv')
df_order_items['shipping_limit_date'] = pd.to_datetime(df_order_items['shipping_limit_date'], format='%d-%m-%Y %H:%M')
```

```python
df_order_items['shipping_limit_date'].dtype
```

'<M8[us]' stands for datetime64[us] in pandas/numpy.

```python
df_order_items.to_csv('olist_order_items_cleaned.csv', index=False)
```

Olist order reviews dataset cleaning

```python
df_order_reviews = pd.read_csv('olist_order_reviews_dataset.csv')
df_order_reviews['review_creation_date'].dtype
```

Olist orders dataset cleaning

```python
df_orders = pd.read_csv('olist_orders_dataset.csv')
```

```python
columns = ["order_purchase_timestamp","order_approved_at","order_delivered_carrier_date",
 "order_delivered_customer_date","order_estimated_delivery_date"]

print(df_orders['order_purchase_timestamp'].dtype, df_orders['order_approved_at'].dtype,
      df_orders['order_delivered_carrier_date'].dtype, df_orders['order_delivered_customer_date'].dtype,
      df_orders['order_estimated_delivery_date'].dtype)
```

```python
df_orders[columns] = df_orders[columns].apply(pd.to_datetime)

print(df_orders['order_purchase_timestamp'].dtype, df_orders['order_approved_at'].dtype,
      df_orders['order_delivered_carrier_date'].dtype, df_orders['order_delivered_customer_date'].dtype,
      df_orders['order_estimated_delivery_date'].dtype)
```

```python
df_orders.to_csv('olist_orders_cleaned.csv', index=False)
```

Olist products dataset cleaning

```python
df_products = pd.read_csv('olist_products_dataset.csv')
translations = pd.read_csv('product_category_name_translation.csv')
```

```python
translations_map = translations.set_index('product_category_name')['product_category_name_english']

df_products['product_category_name_english'] = df_products['product_category_name'].map(translations_map)
```

```python
df_products.to_csv('olist_products_cleaned.csv', index=False)
```
